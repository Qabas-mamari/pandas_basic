### 2. Which country has the most men's gold medals in singles badminton over the years? Sort the result alphabetically by players's names.

<h3 style="color:red">solution 1</h3>

```python
singles_badminton= oo[oo.Sport=='Badminton']
singles_badminton
```

```python
gold = singles_badminton[singles_badminton.Medal =='Gold']
gold
```

```python
men = gold[gold.Gender == 'Men']
men
```

```python
sort = men.NOC.value_counts()
sort
```

<h3 style="color:red">solution 2</h3>

```python
gbm = oo[(oo.Medal=="Gold") & (oo.Sport=="Badminton") & (oo.Gender=="Men")]  
gbm.sort_values(by='Athlete')
```

```python
gbm.NOC.value_counts()
```
